```python
# Graficos normales
import FuncionesG.Barras as DiagramaBarras
import FuncionesG.Pie as DiagramaPastel
import FuncionesG.Histograma as DiagramaHistograma
import FuncionesG.Ojiva as DiagramaOjiva
import FuncionesG.Poligonos as DiagramaPoligono

# GRAFICO con limites superiores e inferiores
import FuncionesG.plotBarras as GraficoBarras
import FuncionesG.plotPie as GraficoPastel
import FuncionesG.plotHistograma as GraficoHistograma
import FuncionesG.plotOjiva as GraficoOjiva
import FuncionesG.plotPoligono as GraficoPoligono

#Funciones 
import FUNCIONES.clases_grouped as ClasesAgrupadas
import FUNCIONES.clases_str_sort as clas_str_sort
import FUNCIONES.fa_grouped as CalcularFABS
import FUNCIONES.format_str as FormatearStrings
import FUNCIONES.frec_abs as abs
import FUNCIONES.frec_acum as FrecAcumulada
import FUNCIONES.frec_rel as FrecRelativa
import FUNCIONES.ordenar_asc as OrdenarArregloAsendente
import FUNCIONES.sort_clases_fa as OrdenarFrecAbsolutas
import FUNCIONES.tabla_grouped as TablaCompleta
import FUNCIONES.tabla as Tabla4Datos
```

```python
# Datos de entrada
datos = ["Pumas", "Am√©rica", "Chivas", "Pumas", "Chivas", "Chivas", "Am√©rica", "Toluca",
          "Toluca", "Pumas", "Santos", "Necaxa", "Am√©rica", "Am√©rica", "Am√©rica", "Pumas",
            "Santos", "Pumas", "Pumas", "Pumas", "Am√©rica", "Chivas", "Am√©rica", "Pumas",
              "Necaxa", "Toluca", "Chivas", "Am√©rica", "Toluca", "Santos" ]

# Formatea los datos para evitar espacios y excepciones en cadena
datos_formated = FormatearStrings.format_str(datos)

# Ordena la lista de datos
datos_sorted = clas_str_sort.clases_str_sort(datos_formated)

# Calcula la frecuencia absoluta de los datos
clases, fa_ab = abs.frec_abs(datos_sorted)

# Crea una copia de las clases originales sin ordenar
clases_orden = clases.copy()

# Obtiene las frecuencias absolutas ordenadas ascendente
fa_sorted = OrdenarFrecAbsolutas.sort_clases_fa(clases_orden, OrdenarArregloAsendente.ordenar_asc(clases), fa_ab)

# Calcula la frecuencia relativa
frecuencia_rel = FrecRelativa.frec_rel(fa_sorted)

# Calcula la frecuencia relativa acumulada
frecuencia_rel_acum = FrecAcumulada.frec_acum(frecuencia_rel)

# Imprime las clases, su frecuencia relativa y su frecuencia relativa acumulada ordenadas de forma ascendente
Tabla4Datos.tabla(clases, fa_sorted, frecuencia_rel, frecuencia_rel_acum)

# Llama a las gr√°ficas para ingresar sus par√°metros

DiagramaBarras.plot_barras(clases, fa_sorted, clases)  # Gr√°fico de barras
DiagramaPastel.plot_pie(fa_sorted, clases)  # Gr√°fico de pie
```
