## Numeros enteros y tabla de 4 datos

```python
# Datos de entrada
datos_entrada = [0, 0, 0, 1, 1, 4, 2, 2, 0, 0, 2, 4, 8, 8, 8, 8, 8, 8, 2, 2, 2, 2, 2, 1, 1, 0, 1, 2, 8, 8]

# Calcula la frecuencia absoluta de los datos
clases, fa_ab = abs.frec_abs(datos_entrada)

# Crea una copia de las clases originales sin ordenar
clases_orig = clases.copy()

# Obtiene las frecuencias absolutas ordenadas ascendente
fa_sorted = OrdenarFrecAbsolutas.sort_clases_fa(clases_orig, OrdenarArregloAsendente.ordenar_asc(clases), fa_ab)

# Calcula la frecuencia relativa
frecuencia_rel = FrecRelativa.frec_rel(fa_sorted)

# Calcula la frecuencia relativa acumulada
frecuencia_rel_acum = FrecAcumulada.frec_acum(frecuencia_rel)

# Imprime las clases, su frecuencia relativa y su frecuencia relativa acumulada ordenadas de forma ascendente
Tabla4Datos.tabla(clases, fa_sorted, frecuencia_rel, frecuencia_rel_acum)
```

## Tabla de 4 datos con palabras

```python
# Datos de entrada
datos = ["toronja", "Cola", "Manzana", " Toronja", "Cola",
         "Toronja", "Cola light", "Cola", "Lim√≥n", "Manzana","Naranja", "Naranja",
         "Cola light", "Lim√≥n", "Cola","Toronja", "Cola", "Lim√≥n", "Toronja", "Cola light",
         "Cola light", "Cola", "Lim√≥n", "Cola light", "Cola","Manzana", "Cola", "Naranja", "Cola", "Cola light",]

# Formatea los datos para evitar espacios y excepciones en cadena
datos_formated = FormatearStrings.format_str(datos)

# Ordena la lista de datos
datos_sorted = clas_str_sort.clases_str_sort(datos_formated)

# Calcula la frecuencia absoluta de los datos
clases, fa_ab = abs.frec_abs(datos_sorted)

# Crea una copia de las clases originales sin ordenar
clases_orden = clases.copy()

# Obtiene las frecuencias absolutas ordenadas ascendente
fa_sorted = OrdenarFrecAbsolutas.sort_clases_fa(clases_orden, OrdenarArregloAsendente.ordenar_asc(clases), fa_ab)

# Calcula la frecuencia relativa
frecuencia_rel = FrecRelativa.frec_rel(fa_sorted)

# Calcula la frecuencia relativa acumulada
frecuencia_rel_acum = FrecAcumulada.frec_acum(frecuencia_rel)

# Imprime las clases, su frecuencia relativa y su frecuencia relativa acumulada ordenadas de forma ascendente
Tabla4Datos.tabla(clases, fa_sorted, frecuencia_rel, frecuencia_rel_acum)

# Llama a las gr√°ficas para ingresar sus par√°metros
DiagramaHistograma.plot_hist(clases, fa_sorted, clases)  # Gr√°fico de histograma
DiagramaBarras.plot_barras(clases, fa_sorted, clases)  # Gr√°fico de barras
DiagramaPastel.plot_pie(fa_sorted, clases)  # Gr√°fico de pie
DiagramaPoligono.plot_poligono(clases, fa_sorted, frecuencia_rel, clases)  # Gr√°fico de pol√≠gono
DiagramaOjiva.plot_ojiva(clases, list(range(1, len(clases)+1)), fa_sorted, [], frecuencia_rel_acum, [])  # Gr√°fico de ojiva

```

## Tabla grande con punto decimal

```python
# Datos de entrada
dt = [0.33, 1.65, 0.99, 1.32, 1.32, 0, 1.65, 0.33, 1.65, 0.66, 0.99, 0.66, 1.65, 0.33, 0.33, 1.32, 0.99, 0.66, 0.66, 0]
# Datos ordenados
dt = OrdenarArregloAsendente.ordenar_asc(dt)

# Obtenemos clases, lim inf, lim. sup, y marcas de clase
clases, lim_inf, lim_sup, mrks = ClasesAgrupadas.clases_grouped(dt)

# Frecuencia absoluta
fa = CalcularFABS.fa_grouped(lim_inf, lim_sup, dt)

# Frecuencia relativa
fr = FrecRelativa.frec_rel(fa)

# Frecuencia relativa acumulada
facum = FrecAcumulada.frec_acum(fr)

# Imprecion en tabla para todos los datos
TablaCompleta.tabla_grouped(clases, lim_inf, lim_sup, mrks, fa, fr, facum)
# Llamada de las funciones para graficar
GraficoHistograma.plot_grafHist(fr, mrks, clases)# Grafico Histograma
GraficoBarras.plot_grafBarras(fr, clases, mrks) # Grafica de barras
GraficoPastel.plot_grafPie(fa, mrks) # Grafico Pie
GraficoPoligono.plot_grafPoligono(clases, mrks, fr) # Poligono de frecuencias
#Si mrks es una lista de marcas de clase, entonces valores_ref_eje = lista de n√∫meros que van desde 1 hasta el n√∫mero de marcas de clase
valores_ref_eje = list(range(1, len(mrks) + 1)) #  Marcas de clase en el eje x 
GraficoOjiva.plot_grafOjiva(mrks, valores_ref_eje, facum) # Grafico de Ojiva
```
